```python
import re
import pandas as pd
import wombat
from wombat.models import dbsession, engine, Item, ItemFeature

features = dbsession.query(ItemFeature).filter(ItemFeature.category == 1).all()
query = "SELECT brand, title, item_type FROM items WHERE brand != 'LENDER SUBMISSION FILL IN' AND rent_per_week < 1000;"
res = engine.execute(query).fetchall()
l = [feature.name for feature in features]
# brands = [res['brand'] for feature in features]

adj_one_hot_df = pd.DataFrame(columns = l).astype('int')
df = adj_one_hot_df
for r in res:
    print(r[1])
```

```python
columns = df.columns.values
row = []
string = "satin chiffon lace sexy diamond dress"
for column in columns:
    match = re.search('{}'.format(column), string, re.IGNORECASE) 
    if match:
        row.append(1)
    else:
        row.append(0)
row_df = pd.DataFrame([row,row], columns)
df.append(row_df)
```
